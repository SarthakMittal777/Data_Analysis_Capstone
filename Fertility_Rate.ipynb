```python
import numpy as np
import pandas as pd

# Define the CSV file path
csv_file_path = r"C:\Users\mitta\Downloads\total_fertility_rate.csv"

# Using NumPy
# Read the CSV file using NumPy's genfromtxt function
# This will create a NumPy array
data_numpy = np.genfromtxt(csv_file_path, delimiter=',', skip_header=1)

# Using Pandas
# Read the CSV file using Pandas' read_csv function
# This will create a DataFrame
df = pd.read_csv(csv_file_path)

# Display the loaded data
print("Data loaded using NumPy:")
print(data_numpy)

print("\nData loaded using Pandas (as 'df'):")
print(df)
```
